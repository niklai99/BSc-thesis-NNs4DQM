```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from dataclasses import dataclass
from typing import Optional
```

## Model class

```python

@dataclass
class ModelClass:

    # model description
    id: int
    batch_norm: bool
    hidden_batch: bool
    custom_output_activation: bool
    output_activation_const: Optional[float] = None

    # model performance
    fitted_chi2: Optional[float]            = None
    chi2: Optional[float]                   = None
    compatibility_1: Optional[float]        = None
    compatibility_2: Optional[float]        = None
    pval: Optional[float]                   = None
    significance: Optional[float]           = None
    median: Optional[float]                 = None
    median_pval: Optional[float]            = None
    median_significance: Optional[float]    = None

    def __repr__(self):
        return_string = (
            f'Model {self.id}\n\n'
            f'Batch Normalization Layer: {self.batch_norm} \n'
            f'Hidden Batch Normalization Layers: {self.hidden_batch} \n'
            f'Custom Output Activation: {self.custom_output_activation} \n'
            f'Activation Constant: {self.output_activation_const} \n'
        )
        return return_string

    def add_performance(
                    self,
                    fitted_chi2: float,
                    chi2: float,
                    compatibility_1: float,
                    compatibility_2: float,
                    pval: float,
                    significance: float,
                    median: float,
                    median_pval: float,
                    median_significance: float
                ):
        self.fitted_chi2          = fitted_chi2
        self.chi2                 = chi2
        self.compatibility_1      = compatibility_1
        self.compatibility_2      = compatibility_2
        self.pval                 = pval
        self.significance         = significance
        self.median               = median
        self.median_pval          = median_pval
        self.median_significance  = median_significance
        
```

## PLOTS

```python

df_performance = df.loc[:, 'fitted_chi2':'median_significance']
df_performance
```

```python

for i in range(len(df_performance.columns)):

    fig, ax = plt.subplots(figsize=(12,7), facecolor='white')

    ax=sns.pointplot(x=df_performance.index, y=df_performance.columns[i], data=df_performance,
                      color='#009cff', join=False, label='model performance')

    if i==0:
        target=10
    elif i==1:
        target=7
    elif i==2:
        target=1
    elif i==3:
        target=0
    elif i==4:
        target=0.5
    elif i==5:
        target=0
    elif i==6:
        target=9.342
    elif i==7:
        target=0.5
    elif i==8:
        target=0
    else:
        target = np.nan

    ax.axhline(target, color='r', linestyle='dashed', linewidth=2, label='target')

    ax.legend()

    ax.set_title(df_performance.columns[i], fontsize=18)

    fig.tight_layout()
    plt.show()
```

```python

residuals_df=pd.DataFrame()

for i in range(len(df_performance.columns)):

    fig, ax = plt.subplots(figsize=(12,7), facecolor='white')

    if i==0:
        target=10
    elif i==1:
        target=7
    elif i==2:
        target=1
    elif i==3:
        target=0
    elif i==4:
        target=0.5
    elif i==5:
        target=0
    elif i==6:
        target=9.342
    elif i==7:
        target=0.5
    elif i==8:
        target=0
    else:
        target=np.nan

    residuals_col = df_performance[df_performance.columns[i]]-target
    residuals_df.insert(loc=i, column=df_performance.columns[i]+' residuals', value=residuals_col)

    ax=sns.pointplot(data=df_performance, x=df_performance.index, y=residuals_col,
                      color='#009cff', join=False, label='model performance')


    ax.axhline(0, color='r', linestyle='dashed', linewidth=2, label='target')

    ax.set_title(df_performance.columns[i]+' residuals', fontsize=18)
    ax.set_ylabel(df_performance.columns[i]+' residuals')

    fig.tight_layout()
    plt.show()
    
```

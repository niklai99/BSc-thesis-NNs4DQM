# Model performance visualization and comparison

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from dataclasses import dataclass
```

## Model class

```python
@dataclass
class Model:
    '''NN model class'''
    chi2_th: int     = 10
    median_th: float = 9.342
    def __init__(
                    self,
                    fitted_chi2: float,
                    chi2: float,
                    compatibility_1: float,
                    compatibility_2: float,
                    pval: float,
                    significance: float,
                    median: float,
                    median_pval: float,
                    median_significance: float
                ):
        '''initialize model performance parameters'''
        self.fitted_chi2          = fitted_chi2
        self.chi2                 = chi2
        self.compatibility_1      = compatibility_1
        self.compatibility_2      = compatibility_2
        self.pval                 = pval
        self.significance         = significance
        self.median               = median
        self.median_pval          = median_pval
        self.median_significance  = median_significance
```

```python
models = [model_4, model_5, model_6, model_7, model_8, model_9, model_10, model_11]
indexes = [f'Model_{n}' for n in range(4, 12)]
columns = [
    'Fitted_Chi2', 
    'Chi2', 
    'Compatibility_1', 
    'Compatibility_2', 
    'P_value', 
    'Z', 
    'Median',
    'Median_P_value',
    'Median_Z'
]
```

```python
fitted_chi2s         = [model.fitted_chi2 for model in models]
chi2s                = [model.chi2 for model in models]
compatibility_1s     = [model.compatibility_1 for model in models]
compatibility_2s     = [model.compatibility_2 for model in models]
pvals                = [model.pval for model in models]
significances        = [model.significance for model in models]
medians              = [model.median for model in models]
median_pvals         = [model.median_pval for model in models]
median_significances = [model.median_significance for model in models]
```

```python
data = np.array([
    fitted_chi2s,
    chi2s, 
    compatibility_1s,
    compatibility_2s,
    pvals,
    significances,
    medians,
    median_pvals,
    median_significances
])
```

```python
df = pd.DataFrame(data.T, index=indexes, columns= columns)
```

```python
df
```

### Model performance

```python
for i in range(len(df.index)+1):
    
    fig, ax = plt.subplots(figsize=(12,7))
    
    ax=sns.pointplot(data=df, x=df.index, y=df.columns[i], 
                      color='#009cff', join=False, label='model performance')
    
    if i==0:
        target=10
    elif i==1:
        target=7
    elif i==2:
        target=1       
    elif i==3:
        target=0
    elif i==4:
        target=0.5
    elif i==5:
        target=0
    elif i==6:
        target=9.342
    elif i==7:
        target=0.5
    elif i==8:
        target=0
        
    ax.axhline(target, color='r', linestyle='dashed', linewidth=2, label='target')
    
    ax.legend()
    
    ax.set_title(df.columns[i], fontsize=18)
    
    fig.tight_layout()
    plt.show()
```

### Residuals from target

```python
residuals_df=pd.DataFrame()

for i in range(len(df.index)+1):
    
    fig, ax = plt.subplots(figsize=(12,7))
    
    if i==0:
        target=10
    elif i==1:
        target=7
    elif i==2:
        target=1       
    elif i==3:
        target=0
    elif i==4:
        target=0.5
    elif i==5:
        target=0
    elif i==6:
        target=9.342
    elif i==7:
        target=0.5
    elif i==8:
        target=0
        
    residuals_col = df[df.columns[i]]-target
    residuals_df.insert(loc=i, column=df.columns[i]+' residuals', value=residuals_col)
    
    ax=sns.pointplot(data=df, x=df.index, y=residuals_col, 
                      color='#009cff', join=False, label='model performance')
    
            
    ax.axhline(0, color='r', linestyle='dashed', linewidth=2, label='target')

    ax.set_title(df.columns[i]+' residuals', fontsize=18)
    ax.set_ylabel(df.columns[i]+' residuals')
    
    fig.tight_layout()
    plt.show()
```
